```python
import os
import tensorflow as tf
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
import matplotlib.pyplot as plt

# Define paths
image_folder = "data"  # Folder containing images
csv_path = "chinese_mnist.csv"

# Load CSV file to get labels and image identifiers
data = pd.read_csv(csv_path)

# Function to get image paths and labels from CSV data
def load_image_paths_and_labels(data, image_folder):
    image_paths = []
    labels = []
    for _, row in data.iterrows():
        # Construct image filename based on `suite_id` and `sample_id`
        image_name = f"input_{row['suite_id']}_{row['sample_id']}.jpg"  # Adjust filename pattern if needed
        image_path = os.path.join(image_folder, image_name)
        if os.path.exists(image_path):
            image_paths.append(image_path)
            labels.append(row['value'])  # Assuming 'value' column is the label
    return image_paths, labels

# Load image paths and labels
image_paths, labels = load_image_paths_and_labels(data, image_folder)

# TensorFlow function to load and preprocess images
def load_image(filepath, label):
    image = tf.io.read_file(filepath)
    image = tf.image.decode_jpeg(image, channels=1)  # Assuming grayscale images
    image = tf.image.resize(image, [64, 64])  # Resize images to 64x64
    image = tf.cast(image, tf.float32) / 255.0  # Normalize pixel values
    return image, label

# Create TensorFlow Dataset from image paths and labels
filepaths_ds = tf.data.Dataset.from_tensor_slices((image_paths, labels))
image_label_ds = filepaths_ds.map(load_image)

# Split the dataset into training and testing
train_size = int(0.8 * len(image_paths))
train_dataset = image_label_ds.take(train_size).batch(32)
test_dataset = image_label_ds.skip(train_size).batch(32)

# Build a CNN model
model = Sequential([
    Conv2D(32, (3, 3), activation="relu", input_shape=(64, 64, 1)),
    MaxPooling2D((2, 2)),
    Dropout(0.25),
    Conv2D(64, (3, 3), activation="relu"),
    MaxPooling2D((2, 2)),
    Dropout(0.25),
    Flatten(),
    Dense(128, activation="relu"),
    Dropout(0.5),
    Dense(15, activation="softmax")  # Assuming there are 15 classes
])

# Compile and train the model
model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
history = model.fit(train_dataset, epochs=10, validation_data=test_dataset)

# Plot training history
plt.plot(history.history["accuracy"], label="Training Accuracy")
plt.plot(history.history["val_accuracy"], label="Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.title("Training and Validation Accuracy")
plt.show()

# Evaluate on the test set
test_loss, test_accuracy = model.evaluate(test_dataset)
print(f"Test Accuracy: {test_accuracy:.2f}")
```
